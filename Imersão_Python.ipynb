```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel('/content/_Tabela de ações Pura.xlsx', sheet_name="Principal")
df_principal.head(10)
```

```python
df_total_acoes = pd.read_excel('/content/_Tabela de ações Pura.xlsx', sheet_name="Total_de_acoes")
df_total_acoes.head(10)
```

```python
df_ticker = pd.read_excel('/content/_Tabela de ações Pura.xlsx', sheet_name="Ticker")
df_ticker.head(10)
```

```python
df_chatgpt= pd.read_excel('/content/_Tabela de ações Pura.xlsx', sheet_name="Chatgpt")
df_ticker.head(10)
```

```python
df_principal = df_principal [["Ativo", "Data", "Último (R$)", "Var. Dia (%)"]].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={"Último (R$)": 'valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
df_principal['Var_pct'] = df_principal['var_dia_pct'] / 100

df_principal['Valor_inicial'] = df_principal['valor_final'] / (df_principal['Var_pct'] + 1)

df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how = 'left') #procv
df_principal
```

```python
df_principal = df_principal.drop(columns= ['Código'])
df_principal
```

```python
df_principal['Variacao_rs'] = (df_principal['valor_final'] - df_principal['Valor_inicial']) * df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format ='{:.2f}'.format #formatar os valores tipo float
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int) #formatar coluna para tipo inteiro
df_principal
```

```python
df_principal = df_principal.rename(columns={"Qtde. Teórica": 'Qtd_teorica'}).copy()
df_principal
```

```python
df_principal["Resultado"] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável') )
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how = 'left') #procv
df_principal = df_principal.drop(columns=['Ticker']) #apagar coluna

df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on ='Nome', right_on = 'Nome da Empresa', how = 'left') #procv

df_principal = df_principal.drop(columns=['Nome da Empresa']) #apagar coluna

df_principal
```

```python
df_principal["Cat_Idade"] = df_principal['Idade (anos)'].apply(lambda x: "Mais De 100 anos" if x > 100 else ("Menos de 50 Anos" if x < 50 else "Entre 50 e 100 anos") )
df_principal
```

```python

# Calculando as estatísticas
maior = df_principal['Variacao_rs'].max()
menor = df_principal['Variacao_rs'].min()
media = df_principal['Variacao_rs'].mean()

media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

# Imprimindo os resultados
print(f"Maior\tR$ {maior:.2f}")
print(f"Menor\tR$ {menor:.2f}")
print(f"Média\tR$ {media:.2f}")
print(f"Media de Quem Subiu\tR$ {media_subiu:.2f}")
print(f"Média de Quem Desceu\tR$ {media_desceu:.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']

df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x = 'Resultado', y = "Variacao_rs", text = 'Variacao_rs', title = 'Variação Reais por Resultado') #Gráficos

fig.show() #mostrar a figura
```

```python
df_analise_faixa_estaria = df_principal.groupby('Cat_Idade').agg({'Variacao_rs': 'sum', "Cat_Idade": 'count'})
df_analise_faixa_estaria

```

```python
df_analise_cat_idade = df_principal.groupby('Cat_Idade')['Variacao_rs'].sum().reset_index()
df_analise_cat_idade

fig = px.bar(df_analise_cat_idade, x='Cat_Idade', y='Variacao_rs', text='Variacao_rs', title='Variação Reais por Categoria de Idade')
fig.show()
```

```python
fig = px.pie(df_analise_segmento, names='Segmento', values='Variacao_rs', title='Variação Reais por Segmento')
fig.show()
```
